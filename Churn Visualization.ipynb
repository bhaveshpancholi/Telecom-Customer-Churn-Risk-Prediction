```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Churn History Dataset.csv')
```

```python
df['Churn Indicator'].unique()
```

```python
sns.countplot(df['Churn Indicator'])
```

```python
#checking data balance
l = len(df[df['Churn Indicator'] == ' True.'])
r = len(df[df['Churn Indicator'] == ' False.'])
print('% of trues in data is {}'.format(round((l/len(df))*100,2)))
print('% of falses in data is {}'.format(round((r/len(df))*100,2)))
```

```python
plt.figure(figsize = (10,10))
a = df[df['Churn Indicator'] == ' True.']
b = df[df['Churn Indicator'] == ' False.']

sns.barplot(x = b['international plan'].unique(),y = b['international plan'].value_counts(),
                color="maroon" ,label = 'churn->False')

sns.barplot(x = a['international plan'].unique(),y = a['international plan'].value_counts(),
                color="coral",label = 'churn->True')
plt.legend()
plt.title('INTERNATIONAL PLAN')
plt.show()
```

If international plan is no, then ratio of churn to not churn is more ie. prospect will be probably churn if he did not have international plan in compare to if international plan is yes.

```python
plt.figure(figsize = (10,8))
a = df[df['Churn Indicator'] == ' True.']
b = df[df['Churn Indicator'] == ' False.']

sns.barplot(x = b['voice mail plan'].unique(),y = b['voice mail plan'].value_counts(),
                color="salmon" ,label = 'churn->False')

sns.barplot(x = a['voice mail plan'].unique(),y = a['voice mail plan'].value_counts(),
                color="crimson",label = 'churn->True')

plt.legend()
plt.title('voice mail plan')
plt.show()
```

```python
plt.figure(figsize = (20,12))
a = df[df['Churn Indicator'] == ' True.']
b = df[df['Churn Indicator'] == ' False.']
sns.countplot(b['state'],color = 'plum',label = 'churn->True')
sns.countplot(a['state'],color = 'darkorchid',label = 'churn->False')
plt.legend()
plt.title('States')
plt.show()
```

```python
plt.figure(figsize = (16,8))
a = df[df['Churn Indicator'] == ' True.']
b = df[df['Churn Indicator'] == ' False.']
sns.countplot(b['area code'],color = 'skyblue',label = 'churn->False')
sns.countplot(a['area code'],color = 'c',label = 'churn->True')
plt.legend()
plt.title('area code')
plt.show()
```

```python
plt.figure(figsize = (18,15))
a = df[df['Churn Indicator'] == ' True.']
b = df[df['Churn Indicator'] == ' False.']
plt.hist(b['number vmail messages'],bins = 10,color = 'plum',edgecolor = 'red',label = 'churn->False')
plt.hist(a['number vmail messages'],bins = 10,color = 'palevioletred',edgecolor = 'red',label = 'churn->True')
plt.legend()
plt.title('number vmail messages')
plt.show()
```

```python
plt.figure(figsize = (10,10))
a = df[df['Churn Indicator'] == ' True.']
b = df[df['Churn Indicator'] == ' False.']
sns.kdeplot(a['total day calls'],shade =True,color = 'red',label = 'churn->True')
sns.kdeplot(b['total day calls'],shade =True,color = 'blue',label = 'churn->False')
plt.legend()
plt.title('total day calls distribution')
plt.show()
```
